```python
users = df["telephone"].unique()
```

```python
for user in users:
    data[user] = {"Date": "", "Add to wl":0, "Add to cart": 0, "# wl": 0, "# cart": 0}
```

```python
for index, row in df.iterrows():
    data[row["telephone"]]["Date"] = "17 Sep 2022"
    if (row["Rating"] == 0.7):
        data[row["telephone"]]["Add to wl"] = 1
        data[row["telephone"]]["# wl"] += 1
    if (row["Rating"] == 0.85):
        data[row["telephone"]]["Add to cart"] = 1
        data[row["telephone"]]["# cart"] += 1
```

```python
for user, userdata in data.items():
    print (str(user) + " ", end ="")
    for values in userdata.values():
        print(str(values) + " ", end = "")
    print()
```
